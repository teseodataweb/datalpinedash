```python
import pandas as pd
df = pd.read_csv('bases/TULANCINGO_24.csv')
df = df[['Occupation','Workforce', 'Monthly Wage', 'lastPeriod']]

df.info()
```

```python
df['lastPeriod'] = df['lastPeriod'].round(2)
df['Monthly Wage'] = df['Monthly Wage'].round(2)

df['diferencia'] = round(df['Monthly Wage'] - df['lastPeriod'],2)
df['diferencia_2'] = df.apply(lambda row: round(row['lastPeriod'] / row['Monthly Wage'],2) if row['Monthly Wage'] != 0 else 0, axis=1)
df['Salario_Variables'] = df[['Monthly Wage', 'lastPeriod']].max(axis=1)
df['volumen'] = df['Salario_Variables'] * df['Workforce']
df['pareja'] = round(df['Salario_Variables'] * 1.7,2)
```

```python
# Función para asignar la categoría en función del salario
def asignar_categoria(salario):
    if 10 <= salario < 5000:
        return "G"
    elif 5000 <= salario < 8000:
        return "F"
    elif 8000 <= salario < 12000:
        return "F1"
    elif 12000 <= salario < 16000:
        return "E"
    elif 16000 <= salario < 20000:
        return "E1"
    elif 20000 <= salario < 25000:
        return "D"
    elif 25000 <= salario < 30000:
        return "D1"
    elif 30000 <= salario < 35000:
        return "C"
    elif 35000 <= salario < 40000:
        return "C1"
    elif 40000 <= salario < 45000:
        return "B"
    elif 45000 <= salario < 50000:
        return "B1"
    elif 50000 <= salario < 55000:
        return "A"
    elif 55000 <= salario < 60000:
        return "A1"
    elif salario > 60000:
        return "S"
    else:
        return None  # En caso de que no cumpla con ningún rango
# Aplicar la función al DataFrame
df['categoria_persona'] = df['Salario_Variables'].apply(asignar_categoria)
df['categoria_pareja'] = df['pareja'].apply(asignar_categoria)
```
