```python
import pandas as pd
df = pd.read_csv('C:/xampp/htdocs/datalpine/resources/db/Scrapining/ciudades/pachuca/clean/febrero_2024.csv')
df['ubicacion'].unique
```

```python
# Define a custom function to ensure we always return a list of 3 elements
def split_location(location):
    parts = location.split(',')
    # Pad the list to ensure it has exactly 3 elements
    parts += [None] * (3 - len(parts))
    return parts[:3]  # Return exactly 3 elements

# Apply the function to the 'ubicacion' column
df[['Colonia', 'Municipio', 'Estado']] = df['ubicacion'].apply(split_location).apply(pd.Series)

# Clean up the data by stripping spaces and filling missing values
df['Colonia'] = df['Colonia'].str.strip().fillna('Sin_colonia')
df['Municipio'] = df['Municipio'].str.strip().fillna('Sin_municipio')
df['Estado'] = df['Estado'].str.strip().fillna('Sin_estado')

# Show the final DataFrame
df[['ubicacion', 'Colonia', 'Municipio', 'Estado']]
```
